```python
import random
import pandas as pd
```

```python
src_ls = []
trg_ls = []
for i in range(1000):
    temp_ls_5 = [random.randint(4, 60) for i in range(4)]
    src_ls.append(temp_ls_5)
    trg_ls.append(temp_ls_5[::-1])

    temp_ls_6 = [random.randint(4, 60) for i in range(5)]
    src_ls.append(temp_ls_6)
    trg_ls.append(temp_ls_6[::-1])

    temp_ls_7 = [random.randint(4, 60) for i in range(6)]
    src_ls.append(temp_ls_7)
    trg_ls.append(temp_ls_7[::-1])

    temp_ls_8 = [random.randint(4, 60) for i in range(7)]
    src_ls.append(temp_ls_8)
    trg_ls.append(temp_ls_8[::-1])

    temp_ls_9 = [random.randint(4, 60) for i in range(8)]
    src_ls.append(temp_ls_9)
    trg_ls.append(temp_ls_9[::-1])

    temp_ls_10 = [random.randint(4, 60) for i in range(9)]
    src_ls.append(temp_ls_10)
    trg_ls.append(temp_ls_10[::-1])
```

```python
df = pd.DataFrame({"source": src_ls, "target": trg_ls})
df.to_csv('num_dataset.csv', index=False)
```
